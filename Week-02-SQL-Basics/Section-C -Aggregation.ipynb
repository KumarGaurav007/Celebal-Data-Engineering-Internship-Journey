## Overview

This section focuses on **Aggregate Functions** and **Grouping** in SQL. Aggregate functions are used to summarize data by performing calculations on multiple rows and returning a single result.

These operations are fundamental in analytics, business intelligence, reporting dashboards, and decision-making systems. They help organizations answer business questions such as total sales, average revenue, highest-selling products, customer distribution, and order statistics.

### Q13. Count the total number of orders in the orders table.
**SQL Query**

```python
%%sql
SELECT COUNT(*) AS total_orders 
FROM orders;
```

### Q14. Find the total revenue (SUM of total_amount) from all 'Delivered' orders.
**SQL Query**

```python
%%sql
SELECT status, SUM(total_amount) AS total_revenue 
FROM orders 
WHERE status = 'Delivered' 
GROUP BY status;
```

### Q15. Calculate the average unit_price of products in each category. 
**SQL Query**

```python
%%sql
SELECT category, AVG(unit_price) AS average_price 
FROM products 
GROUP BY category;
```

### Q16. For each order status, find the count of orders and the total revenue. Sort the result by total revenue in descending order. 
**SQL Query**

```python
%%sql
SELECT 
    status, 
    COUNT(order_id) AS total_order, 
    SUM(total_amount) AS total_revenue
FROM orders 
GROUP BY status
ORDER BY total_revenue DESC;
```

### Q17. Find the most expensive (MAX) and cheapest (MIN) product in each category. 
**SQL Query**

```python
%%sql
SELECT 
    category,
    MAX(unit_price) AS most_expensive_price,
    MIN(unit_price) AS cheapest_price
FROM products
GROUP BY category;
```

### Q18. List all product categories where the average unit_price is greater than ₹2000. (Hint: Use HAVING clause) 
**SQL Query**

```python
%%sql
SELECT 
    category,
    AVG(unit_price) AS average_price
FROM products
GROUP BY category
HAVING AVG(unit_price) > 2000;
```
