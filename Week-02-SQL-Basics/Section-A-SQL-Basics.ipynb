### Q4. Identify the Primary Key of each table in the schema.

#### Primary Keys in the Database

| Table | Primary Key |
|--------|-------------|
| customers | `customer_id` |
| products | `product_id` |
| orders | `order_id` |
| order_items | `item_id` |

---

### Why must a Primary Key be UNIQUE?

A Primary Key uniquely identifies every record in a table. No two rows can have the same Primary Key value. This ensures that each record can be accessed, updated, or deleted without ambiguity.

**Example:**

If two customers had the same `customer_id`, the database would not be able to determine which customer a query refers to.

---

### Why must a Primary Key be NOT NULL?

A Primary Key cannot contain `NULL` values because every record must have a valid identifier. A `NULL` value represents an unknown or missing value, making it impossible to uniquely identify a row.

**Example:**

If an order had a `NULL` value for `order_id`, the database would not be able to uniquely identify that order.

---

### Conclusion

A Primary Key is one of the most important constraints in a relational database. It guarantees that every record has a unique and valid identifier, ensuring data integrity and enabling reliable relationships between tables through Foreign Keys.

> **Key Takeaway**
>
> A Primary Key ensures **Entity Integrity** by making every record unique and preventing NULL values. It also serves as the reference point for establishing relationships with other tables using Foreign Keys.

### Q5. What constraints are applied to the `email` column in the `customers` table? What would happen if you tried to insert a duplicate email?

#### Constraints Applied to the `email` Column

The `email` column in the **customers** table has the following constraints:

1. **UNIQUE**
   - Ensures that every customer has a unique email address.
   - No two customers can have the same email.

2. **NOT NULL**
   - Ensures that every customer record must contain an email address.
   - Empty or `NULL` values are not allowed.

---

### What happens if a duplicate email is inserted?

If an attempt is made to insert a customer with an email address that already exists in the table, the database will reject the operation and generate a **UNIQUE constraint violation** error.

This prevents duplicate customer records and helps maintain data integrity.

---

### Example

```sql
INSERT INTO customers
VALUES (
109,
'Rahul',
'Verma',
'aarav.s@email.com',
'Lucknow',
'Uttar Pradesh',
'2024-09-10',
TRUE
);
```

---

### Expected Result

The above query will fail because the email **`aarav.s@email.com`** already exists in the table.

Example error:

```text
ERROR 1062 (23000):
Duplicate entry 'aarav.s@email.com'
for key 'customers.email'
```

---

**SQL Query**

````python
%%sql
INSERT INTO customers
VALUES (
109,
'Rahul',
'Verma',
'aarav.s@email.com',
'Lucknow',
'Uttar Pradesh',
'2024-09-10',
TRUE
);
````

### Conclusion

From the above query and its output we can conclude that the combination of **UNIQUE** and **NOT NULL** constraints ensures that every customer has a valid and distinct email address. These constraints help maintain data quality, prevent duplicate records, and support reliable customer identification.

### Q6. Try inserting a product with `unit_price = -50`. What happens and which constraint prevents it?

#### Objective

The objective of this task is to understand how the **CHECK constraint** protects the database from storing invalid data.

---

### Expected Result

The query will **fail** because the value of `unit_price` is negative.

The `products` table defines the following CHECK constraint:

```sql
CHECK (unit_price > 0)
```

Since **-50** is less than **0**, the database rejects the INSERT statement.

Example error:

```text
ERROR:
Check constraint 'products_chk_unit_price' is violated.
```

### SQL Query

````python
%%sql
INSERT INTO products
(
    product_id,
    product_name,
    category,
    brand,
    unit_price,
    stock_qty
)
VALUES
(
    209,
    'Gaming Mouse',
    'Electronics',
    'Logitech',
    -50,
    100
);
````

### Explanation

A **CHECK constraint** ensures that data inserted into a column satisfies a specified condition.

In this table:

- `unit_price` must always be greater than **0**.
- Negative prices are considered invalid business data.
- Therefore, the database prevents the record from being inserted.

This helps maintain **data accuracy** and **business integrity** by ensuring only valid product prices are stored.

---

### Conclusion

The **CHECK (unit_price > 0)** constraint prevents invalid product prices from being inserted into the database. It acts as a validation rule at the database level, ensuring that every product has a positive selling price.
