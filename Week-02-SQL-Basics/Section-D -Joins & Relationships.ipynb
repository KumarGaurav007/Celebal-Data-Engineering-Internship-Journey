### Q19. Write an INNER JOIN query to display each order along with the customer's first_name and last_name. Show: `order_id`, `order_date`, `first_name`, `last_name`, `total_amount`.

#### Objective

The objective of this query is to combine data from the **orders** and **customers** tables using an **INNER JOIN**. This allows us to display order details together with the customer who placed each order.

---

### Relationship Used

```
customers
-----------
customer_id (PK)
        │
        │
        ▼
orders
-----------
customer_id (FK)
```

The `customer_id` in the **orders** table is a **Foreign Key** that references the `customer_id` in the **customers** table.

---

### SQL Query

````python
%%sql
SELECT
    o.order_id,
    o.order_date,
    c.first_name,
    c.last_name,
    o.total_amount
FROM orders AS o
INNER JOIN customers AS c
ON o.customer_id = c.customer_id;
````

### Explanation

- `INNER JOIN` combines rows from both tables where the `customer_id` matches.
- The alias `o` represents the **orders** table.
- The alias `c` represents the **customers** table.
- Only records with matching `customer_id` values in both tables are returned.

---

### Why INNER JOIN?

An **INNER JOIN** is used because we only want orders that have a valid customer associated with them.

If an order does not have a matching customer record, it will **not** appear in the result.

If the requirement were to display **all customers**, including those who have never placed an order, a **LEFT JOIN** would be the appropriate choice.

---

### Conclusion

The `INNER JOIN` retrieves only those records where a matching `customer_id` exists in both the **customers** and **orders** tables. It is the most commonly used join for combining related data across tables in relational databases.

### Q20. Using a LEFT JOIN, list **ALL** customers and their orders (if any). Customers with no orders should still appear with `NULL` values for order columns.

#### Objective

The objective of this query is to retrieve **all customers**, regardless of whether they have placed an order. If a customer has not placed any orders, the customer details should still be displayed, while the order-related columns should contain `NULL` values.

---

### Relationship Used

```
customers
-----------
customer_id (PK)
        │
        │
        ▼
orders
-----------
customer_id (FK)
```

The `customer_id` in the **orders** table is a **Foreign Key** that references the `customer_id` in the **customers** table.

---

### SQL Query

````python
%%sql
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    o.order_id,
    o.order_date,
    o.total_amount
FROM customers AS c
LEFT JOIN orders AS o
ON c.customer_id = o.customer_id;
````


### Explanation

- `LEFT JOIN` returns **all records** from the left table (`customers`).
- Matching records from the `orders` table are included whenever they exist.
- If a customer has not placed any order, the columns from the `orders` table are displayed as `NULL`.
- This ensures that every customer appears in the result set.

---

### INNER JOIN vs LEFT JOIN

| INNER JOIN | LEFT JOIN |
|------------|-----------|
| Returns only matching records | Returns all records from the left table |
| Non-matching rows are excluded | Non-matching rows are included with `NULL` values |
| Used when only related data is required | Used when complete data from the left table is required |

---

### Conclusion

A **LEFT JOIN** ensures that every customer is included in the result set, even if there is no matching record in the **orders** table. When no matching order exists, the corresponding order columns contain `NULL` values.

### Q21. Write a query using JOINs across three tables (`orders → order_items → products`) to show: `order_id`, `product_name`, `quantity`, `unit_price`, and `discount_pct` for each order item.

#### Objective

The objective of this query is to retrieve complete information about every product included in an order by combining data from three related tables.

This demonstrates how SQL JOINs are used to combine normalized data across multiple tables in a relational database.

---

### Relationship Used

```
orders
---------
order_id (PK)
     │
     │
     ▼
order_items
---------
order_id (FK)
product_id (FK)
     │
     │
     ▼
products
---------
product_id (PK)
```

- `orders.order_id` → `order_items.order_id`
- `products.product_id` → `order_items.product_id`

---

### SQL Query

````python
%%sql
SELECT
    o.order_id,
    p.product_name,
    oi.quantity,
    oi.unit_price,
    oi.discount_pct
FROM orders AS o
INNER JOIN order_items AS oi
    ON o.order_id = oi.order_id
INNER JOIN products AS p
    ON oi.product_id = p.product_id;
````

### Explanation

This query uses **two INNER JOINs** to combine three related tables.

1. The first `INNER JOIN` connects the **orders** table with the **order_items** table using `order_id`.
2. The second `INNER JOIN` connects the **order_items** table with the **products** table using `product_id`.

As a result, each row represents one product purchased in an order along with its quantity, selling price, and discount percentage.

---

### Why Table Aliases?

| Alias | Table |
|--------|-------|
| `o` | orders |
| `oi` | order_items |
| `p` | products |

Table aliases make SQL queries shorter, easier to read, and improve maintainability, especially when working with multiple joined tables.

---

### Conclusion

Using multiple `INNER JOIN`s allows data stored in different normalized tables to be combined into a single meaningful result. This approach minimizes data redundancy while enabling comprehensive business reporting.

### Q22. Explain the difference between `LEFT JOIN` and `RIGHT JOIN` with an example from this schema. When would you use a `FULL OUTER JOIN`?

#### Objective

The objective of this question is to understand the differences between `LEFT JOIN`, `RIGHT JOIN`, and `FULL OUTER JOIN`, and identify when each type of join should be used in real-world database scenarios.

---

## LEFT JOIN

A **LEFT JOIN** returns **all records from the left table** and only the matching records from the right table.

If no matching record exists in the right table, the corresponding columns contain **NULL** values.

### Example


````python
%%sql
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    o.order_id
FROM customers AS c
LEFT JOIN orders AS o
ON c.customer_id = o.customer_id;
````

### Result

- Every customer appears in the output.
- Customers who have not placed an order still appear.
- Their `order_id` is displayed as `NULL`.

---

## RIGHT JOIN

A **RIGHT JOIN** returns **all records from the right table** and only the matching records from the left table.

If no matching record exists in the left table, the corresponding columns contain **NULL** values.

### Example


````python
%%sql
SELECT
    c.customer_id,
    c.first_name,
    o.order_id,
    o.order_date
FROM customers AS c
RIGHT JOIN orders AS o
ON c.customer_id = o.customer_id;
````

### Result

- Every order appears in the output.
- If an order has no matching customer (rare due to Foreign Key constraints), the customer columns will contain `NULL`.

---

## FULL OUTER JOIN

A **FULL OUTER JOIN** returns:

- All rows from the left table
- All rows from the right table
- Matching rows are combined
- Non-matching rows from either table are filled with `NULL`

### When would you use it?

A FULL OUTER JOIN is useful when you need to identify **all matching and non-matching records** from both tables.

Examples include:

- Customers without orders
- Orders without customers
- Data validation
- Data migration
- Data reconciliation between two systems

---

## Note for MySQL

MySQL **does not support `FULL OUTER JOIN` directly.**

It can be simulated by combining a `LEFT JOIN` and a `RIGHT JOIN` using the `UNION` operator.

Example:

````python
%%sql
SELECT 
    customers.customer_id, 
    customers.first_name, 
    orders.order_id, 
    orders.total_amount
FROM customers
LEFT JOIN orders
ON customers.customer_id = orders.customer_id

UNION

SELECT 
    customers.customer_id, 
    customers.first_name, 
    orders.order_id, 
    orders.total_amount
FROM customers
RIGHT JOIN orders
ON customers.customer_id = orders.customer_id;

````

## Comparison

| LEFT JOIN | RIGHT JOIN | FULL OUTER JOIN |
|------------|------------|-----------------|
| Returns all rows from the left table | Returns all rows from the right table | Returns all rows from both tables |
| Non-matching right rows become NULL | Non-matching left rows become NULL | Non-matching rows from both sides become NULL |
| Most commonly used | Less commonly used | Mainly used for reconciliation and auditing |

---

## Conclusion

Choosing the correct JOIN depends on the business requirement.

- Use **LEFT JOIN** when every record from the left table must be preserved.
- Use **RIGHT JOIN** when every record from the right table is required.
- Use **FULL OUTER JOIN** when a complete comparison of both tables is needed, including all matching and non-matching records.

### Q23. Identify all Foreign Key relationships in the schema. Explain what would happen if you tried to insert an order with `customer_id = 999` (which doesn't exist in `customers`).

#### Objective

The objective of this question is to understand how **Foreign Keys** establish relationships between tables and maintain **Referential Integrity** in a relational database.

---

## Foreign Key Relationships in the Database

The database contains the following Foreign Key relationships:

| Parent Table | Child Table | Foreign Key |
|--------------|-------------|-------------|
| customers | orders | `orders.customer_id` → `customers.customer_id` |
| orders | order_items | `order_items.order_id` → `orders.order_id` |
| products | order_items | `order_items.product_id` → `products.product_id` |

These relationships ensure that every child record references a valid record in its parent table.

---

## Relationship Diagram

```

│
▼
orders
(customer_id)

orders
(order_id)
│
│
▼
order_items
(order_id)

products
(product_id)
│
│
▼
order_items
(product_id)

```

---

## Example Query

Suppose we try to insert an order for a customer who does not exist.

````python
%%sql
INSERT INTO orders
(
    order_id,
    customer_id,
    order_date,
    status,
    total_amount
)
VALUES
(
    1011,
    999,
    '2024-07-15',
    'Pending',
    2500.00
);
````

## What will happen?

The query will **fail** because there is no customer with `customer_id = 999` in the `customers` table.

Since `orders.customer_id` is defined as a **Foreign Key**, the database verifies whether the referenced customer exists before allowing the record to be inserted.

---

## Expected Error

```text
ERROR 1452 (23000):
Cannot add or update a child row:
a foreign key constraint fails
```
---

## Why does this happen?

A Foreign Key enforces **Referential Integrity**.

This means:

- Every order must belong to a valid customer.
- The database prevents orphan records.
- Invalid relationships cannot be created.

Without this constraint, orders could exist without corresponding customers, leading to inconsistent and unreliable data.

---

### Primary Key vs Foreign Key

| Primary Key | Foreign Key |
|-------------|-------------|
| Uniquely identifies a record | References a Primary Key in another table |
| Must be UNIQUE and NOT NULL | Can contain duplicate values (and sometimes NULL, depending on the design) |
| One per table | Multiple Foreign Keys can exist in a table |
| Ensures Entity Integrity | Ensures Referential Integrity |

---

## Conclusion

Foreign Keys maintain relationships between tables by ensuring that every referenced value exists in the parent table. Attempting to insert an order with a non-existent `customer_id` violates the Foreign Key constraint, and the database rejects the operation to preserve data integrity.
