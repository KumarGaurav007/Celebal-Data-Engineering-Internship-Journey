### Discount Column Analysis

```python
df["discount"].fillna(0, inplace=True)
```

### Seller Information Analysis

```python
df["seller_name"].fillna("Unknown", inplace=True)
```

```python
df["seller_information"].fillna("Not Available", inplace=True)
```

```python
df["videos"].fillna("No Video", inplace=True)
```

### What customer said

```python
df["what_customers_said"].fillna(
    "No Customer Review",
    inplace=True
)
```

### Variations

```python
df["variations"].fillna(
    "No Variations",
    inplace=True
)
```

# Data Type Conversion

```python
df["final_price"] = (
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["final_price"].head()
```

# Derived Column

```python
df["discount_amount"] = (
    df["initial_price"]
    - df["final_price"]
)
```

# Filtering Data
### Highly Rated Products

```python
high_rated = df[df["rating"] > 4]

high_rated.head()
```

### Column Selection

```python
df[
    [
        "title",
        "rating",
        "initial_price",
        "final_price",
        "category"
    ]
].head()
```

# Save Cleaned Dataset

```python
df.to_csv(
    "cleaned-data/cleaned_shopping_data.csv",
    index=False
)
```
